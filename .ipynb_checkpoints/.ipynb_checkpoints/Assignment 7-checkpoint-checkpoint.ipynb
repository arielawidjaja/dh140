```python
response = requests.get('https://www.gutenberg.org/cache/epub/1112/pg1112.txt')
```

```python
response.text
```

```python
wordlist = response.text
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(wordlist)
```

```python
#tokenize wordlist
tokenized_sents = [word_tokenize(i) for i in wordlist]
for i in tokenized_sents:
    print (i)
```

```python
#make the wordlist
wordlist = []
for s in sent:
    for w in word_tokenize(s):
        wordlist.append(w)
```

```python
print(wordlist)
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
wordlistnostop = [w for w in wordlist if w not in myStopWords]
```

```python
print(wordlistnostop)
```

```python
from nltk.stem.lancaster import LancasterStemmer
```

```python
wordLancasterStems = [LancasterStemmer().stem(w) for w in wordlistnostop]
```

```python
print(wordLancasterStems)
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordlistnostop)
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#turning the frequency stuff into a df for easier statistical calculation
wordlist = []
wordfreq = []
for i in sorted(freq, key=freq.get, reverse=True):
        wordlist.append(i)
        wordfreq.append(freq[i])
df = pd.DataFrame({'wordlist':wordlist,'wordfreq':wordfreq})
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
result.corr()
```

```python
#remember to plot result and not "df"
result.plot(kind='scatter',
        x='bmi',
        y='target');
```

```python
#use df column instead of defining as string, add .values since df does not come with reshape 
X_new = features_df[['bmi']].values.reshape(-1,1)
y = target_df
```

```python
reg = LinearRegression().fit(X_new, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X_new
```

```python
plt.plot(X_new,y,'ro',X_new,ytrain,'b-');
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```
