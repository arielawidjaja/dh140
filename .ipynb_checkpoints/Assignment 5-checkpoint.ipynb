```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://corgis-edu.github.io/corgis/csv/')
```

```python
response.text
```

```python
html_string = response.text
```

```python
document = BeautifulSoup(html_string, "html.parser")
```

```python
#find only does one result, use find_all
document.find_all('h3')
```

```python
list = []
```

```python
#to find text without "h3"
for i in document.find_all('h3'):
    list.append(i.text.strip())
```

```python
response = requests.get('https://corgis-edu.github.io/corgis/csv/cars/')
```

```python
response
```

```python
response.text
```

```python
html_string = response.text
```

```python
document = BeautifulSoup(html_string, "html.parser")
```

```python
document.find_all('a')
```

```python
#show download link only if available
for i in document.find_all("a"):
    if i.has_attr('download'):
        print(i.text.strip() + ' : accessible at https://corgis-edu.github.io/corgis/csv/cars/' + i.attrs['href'])
```

```python
def corgisdf():
    pd.read_csv('https://corgis-edu.github.io/corgis/csv/cars/../../datasets/csv/cars/cars.csv')
    return corgisdf
```

```python
corgisdf = pd.read_csv('https://corgis-edu.github.io/corgis/csv/cars/../../datasets/csv/cars/cars.csv')
```

```python
corgisdf[:15].sort_values(by='Dimensions.Height',ascending=True).plot(kind='barh',
             x='Identification.Make',
             y='Dimensions.Height')
```
